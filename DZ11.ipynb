```python
x = sympy.Symbol('x')
expr = trigsimp(sin(x)**2 - cos(x)**2)
expr
```

```python
expr_solve = sp.solve(expr)
expr_solve
```

```python
expr_diff = sp.diff(expr)
expr_diff
```

```python
graph=sp.plot(expr)
```

```python
rise_exp=sp.solve(expr_diff>0)
rise_exp
```

```python
fall_exp=sp.solve(expr_diff<0)
fall_exp
```
