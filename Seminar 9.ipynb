```python
import sympy as sp
```

```python
x = sp.Symbol('x')
expr = 5*x**2 + 10*x - 30
expr
```

```python
expr_solve = sp.solve(expr)
expr_solve
```

```python
expr_diff = sp.diff(expr)
expr_diff
```

```python
graph=sp.plot(expr)
```

```python
rise_exp=sp.solve(expr_diff>0)
rise_exp
```

```python
fall_exp=sp.solve(expr_diff<0)
fall_exp
```

```python
peak_exp=sp.solve(expr_diff)
peak_exp
```

```python
peak_exp_origin=sp.solve(expr>0)
peak_exp_origin
```

```python
fall_expr_origin=sp.solve(expr<0)
fall_expr_origin
```
